```python
import json
import cv2 as cv
```

```python
with open("C:\\Users\\Artem\\mipt2024s-5-Tiniakov-A-D\\data\\fst_50_barcodes.json") as fin:
    data = json.load(fin)
    for key, value in data.items():
        with open("C:\\Users\\Artem\\yolo_barcode_dataset\\Training\\labels\\"+value['filename'][:-4]+".txt", 'w') as fout:
            width, height, _ = cv.imread("C:\\Users\\Artem\\mipt2024s-5-Tiniakov-A-D\\data\\images\\"+value['filename']).shape
            for region in value['regions']:
                if region['region_attributes']['barcode'] == "1d":
                    fout.write("0 ")
                else:
                    fout.write("1 ")
                for x, y in zip(region['shape_attributes']['all_points_x'], region['shape_attributes']['all_points_y']):
                    fout.write(f"{x/width} {y/height} ")
                fout.write('\n')
```
