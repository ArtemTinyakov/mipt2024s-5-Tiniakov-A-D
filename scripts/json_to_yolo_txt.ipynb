```python
import json
import cv2 as cv
import os
```

```python
data_path = "C:/Users/Artem/Downloads/kulakovaanna_mipt2024s-5-kulakova-a-d_main_data-barcode_dataset/"

for filename in os.listdir(data_path + "labels/"):
    if filename[-4:] == "json":
        with open(data_path + "labels/" + filename) as fin:
            data = json.load(fin)
            for key, value in data["_via_img_metadata"].items():
                with open("C:/Users/Artem/yolo_barcode_dataset/Training/labels/"+value['filename'][:-4]+".txt", 'w') as fout:
                    height, width, _ = cv.imread(data_path + "images/" + value['filename']).shape
                    for region in value['regions']:
                        if region['region_attributes']['barcode'] == "1d":
                            fout.write("0 ")
                        else:
                            fout.write("1 ")
                        for x, y in zip(region['shape_attributes']['all_points_x'], region['shape_attributes']['all_points_y']):
                            fout.write(f"{x/width} {y/height} ")
                        fout.write('\n')
```

```python
# just check that labeling is ok

import matplotlib.pyplot as plt

img = cv.imread("C:/Users/Artem/yolo_barcode_dataset/Training/images/IMG_1002.jpg")
height, width, _ = img.shape

with open("C:/Users/Artem/yolo_barcode_dataset/Training/labels/IMG_1002.txt", 'r') as fin:
    data = fin.readline().split(' ')[1:-1]
    x = [float(data[i*2])*width for i in range(int(len(data)/2))]
    y = [float(data[i*2+1])*height for i in range(int(len(data)/2))]

print(x, y, img.shape)

plt.imshow(img)
plt.scatter(x, y)
plt.show()
```
