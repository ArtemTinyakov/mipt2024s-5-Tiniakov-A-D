```python
from pyzbar.pyzbar import decode, ZBarSymbol
import cv2 as cv
from ultralytics import YOLO
import os
import numpy as np

data_path = "C:/Users/Artem/bboxed_barcodes_dataset/Validation/images/"
images = os.listdir(data_path)
output = "C:/Users/Artem/mipt2024s-5-Tiniakov-A-D/opencv_pyzbar_yolo_pre_tests/"
```

```python
opencv_barcode_detector = cv.barcode.BarcodeDetector('sr.prototxt', 'sr.caffemodel')
yolo_model = YOLO("C:/Users/Artem/runs/segment/train/weights/best.pt")
```

```python
from copy import deepcopy

for image_name in images:
    img = cv.imread(data_path + image_name)
    
    pyzbar_predict = decode(img)
    pyzbar_sample = deepcopy(img)
    
    if len(pyzbar_predict) > 0:
        for predict in pyzbar_predict:
            pts = [[predict.polygon[i].x, predict.polygon[i].y] for i in range(len(predict.polygon))]
            pts = np.array(pts)
            pts = pts.reshape((-1,1,2))
            cv.polylines(pyzbar_sample, [pts], True, (0,0,255), 10)
    cv.imwrite(output + "pyzbar/" + image_name, pyzbar_sample)

    opencv_retval, opencv_points = opencv_barcode_detector.detect(img)
    opencv_sample = deepcopy(img)

    if not opencv_points is None:
        for pts in opencv_points:
            pts = np.array(pts)
            pts = pts.reshape((-1,1,2))
            pts = pts.astype(np.int32)
            cv.polylines(opencv_sample, [pts], True, (0,0,255), 10)
    cv.imwrite(output + "opencv/" + image_name, opencv_sample)
        
    
    yolo_predict = yolo_model(img, save=True)
    os.replace("C:/Users/Artem/runs/segment/predict/image0.jpg", output + "yolo/" + image_name)
```
